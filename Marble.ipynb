```python
from PIL import Image
import os

# Folder path containing the original images
input_folder_path = '/content/drive/MyDrive/Products'

# Output folder path where resized images will be saved
output_folder_path = '/content/drive/MyDrive/ResizedProducts'
os.makedirs(output_folder_path, exist_ok=True)  # Ensure the output directory exists

# New dimensions for resizing
new_width = 1300
new_height = 1300

# Iterate over each file in the folder
for filename in os.listdir(input_folder_path):
    # Construct the full path to the image file
    input_image_path = os.path.join(input_folder_path, filename)
    output_image_path = os.path.join(output_folder_path, filename)

    # Open the image using PIL
    try:
        with Image.open(input_image_path) as img:
            # Resize the image
            resized_img = img.resize((new_width, new_height))

            # Save the resized image to the output folder
            resized_img.save(output_image_path)

            # Print confirmation message
            print(f"Resized image '{filename}' saved successfully.")
    except FileNotFoundError:
        print(f"The specified image file '{filename}' was not found.")
    except Exception as e:
        print("An error occurred:", e)
```

```python
from PIL import Image
import os

# Path to the folder containing background images
background_folder_path = '/content/drive/MyDrive/Backgrounds/P1'

# Path to the masked image
masked_image_path = '/content/drive/MyDrive/Masked/masked_1.jpg'

# Path to the output folder where the modified images will be saved
output_folder_path = '/content/drive/MyDrive/Outputs/P1'

# Iterate through each background image in the folder
for filename in os.listdir(background_folder_path):
    if filename.endswith(".jpg") or filename.endswith(".jpeg"):
        # Open the background image
        background_image_path = os.path.join(background_folder_path, filename)
        background_image = Image.open(background_image_path)

        # Resize the background image to match the size of the masked image
        masked_img = Image.open(masked_image_path)
        background_image_resized = background_image.resize((masked_img.width, masked_img.height))

        # Paste the masked image onto the resized background image
        background_image_resized.paste(masked_img, (0, 0), masked_img)

        # Save the final image
        output_image_path = os.path.join(output_folder_path, f"{os.path.splitext(filename)[0]}_masked.jpg")
        background_image_resized.save(output_image_path, format='JPEG')
```
